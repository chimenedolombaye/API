```python
# Importation des bibliothèques nécessaires
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud
```

```python
# Chargement des données
data = pd.read_csv("creditcard.csv")
data.head(5)
```

```python
data.info()
```

```python
# Analyse approfondie des caractéristiques
data.describe()
```

```python
data.isna().sum()
```

```python
# Affichez la liste des colonnes dans votre ensemble de données
print(data.columns)
```

```python
# Visualisation des données
plt.figure(figsize=(10, 8))#Cette ligne crée une  figure avec une taille de 10 pouces de largeur et 8 pouces de hauteur
sns.countplot(x="Class", data=data)#Cette ligne utilise la fonction countplot de Seaborn pour créer un diagramme à barres.
plt.title("Distribution des classes (Fraude: 1, Non-Fraude: 0)")
plt.show()
```

```python
# Échantillonnage aléatoire de 1000 points de données
data_sample = data.sample(n=1000, random_state=42)

# Visualisation des transactions frauduleuses et non frauduleuses en fonction du temps
plt.figure(figsize=(12, 8))
sns.countplot(x="Time", hue="Class", data=data_sample, palette="Set1")
plt.title("Distribution des transactions au fil du temps (Fraude: 1, Non-Fraude: 0)")
plt.xlabel("Temps")
plt.ylabel("Nombre de transactions")
plt.legend(title="Classe", labels=["Non-Fraude", "Fraude"])
plt.show()
```

```python
# Matrice de corrélation
correlation_matrix = data.corr()

# Visualisation de la matrice de corrélation avec seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de Corrélation")
plt.show()

```

```python
# Supprimer les lignes avec des valeurs manquantes
data = data.dropna()
```

```python
# Identifier les valeurs aberrantes avec une boîte à moustaches (pour les variables numériques)
numerical_features = data.select_dtypes(include=['float64', 'int64']).columns
plt.figure(figsize=(12, 8))
data[numerical_features].boxplot()
plt.title("Boîte à Moustaches pour la Détection des Valeurs Aberrantes")
plt.show()
```
